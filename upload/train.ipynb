## Step 3 — Train

This cell runs `scripts/03_train.py` — the **exact same** training script used
locally. Changes to `03_train.py` are automatically reflected here via
`prepare_colab.py`.

```python
import os, subprocess, sys

DRIVE_ROOT   = '/content/drive/MyDrive/ballnet/v5'
UPLOAD_DIR   = os.path.join(DRIVE_ROOT, 'upload')
DOWNLOAD_DIR = os.path.join(DRIVE_ROOT, 'clean_download')

os.makedirs(DOWNLOAD_DIR, exist_ok=True)

# Use Popen to manually pipe the output
process = subprocess.Popen(
    ['python', '-u', 'scripts/03_train.py'],
    cwd=UPLOAD_DIR,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT, # Merges stderr into stdout
    text=True
)

# Stream the output line-by-line
for line in process.stdout:
    print(line, end='')

process.wait()

print(f'\nExit code: {process.returncode}')
print(f'Model weights saved to: {DOWNLOAD_DIR}/')
```
